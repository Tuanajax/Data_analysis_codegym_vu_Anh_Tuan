```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
df = pd.read_csv('OnlineRetail.csv',encoding = "ISO-8859-1")
df.shape
```

```python
df.info()
```

```python
# Kiểm tra khuyết thiếu
df.isna()
```

```python
# In ra dữ liệu ngoại lai
df[df['Quantity']<0]
```

```python
# Xóa bỏ dữ liệu ngoại lai
df = df[df['Quantity']>=0]
df.shape
```

```python
# Xóa bỏ dòng dữ liệu bị khuyết
df1 = df.dropna()
df1.shape
```

```python
# xóa toàn bộ dòng bị khuyết thiếu
df2 = df.dropna(how='all')
df2.shape
```

```python
# Giữ lại những dòng có ít nhất 7 giá trị khuyết thiếu
df3 = df.dropna(thresh=7)
df3.shape
```

```python
# Xóa những hàng có gia strị khuyết thiếu trên cột customerID
df4= df.dropna(subset=['CustomerID'])
df4.shape
```

```python
# Thay thế giá trị khuyết trên cốh
df5 =df
df5['CustomerID'] =df['CustomerID'].fillna(-1)
df5[df5['CustomerID']==-1]
```

```python
# Thay thế giá trị khuyết ở cột Country bằng gia trị trước đó
df5['Country'] = df['Country'].fillna(method='ffill')
df5
```

```python
# Xử lý ngoại lai ở cột quantity
sns.boxplot(x=df5['Quantity'])
```

```python
# Xóa dữ liệu ngoại lai
Q1 = df1['Quantity'].quantile(0.25)
Q3 = df1['Quantity'].quantile(0.75)
IQR = Q3 - Q1
# xác định phần tử không phải ngoại lai
df6 = df1
df6['outlier'] = ~((df1['Quantity'] < (Q1 - 1.5*IQR)) | (df1['Quantity'] > (Q3 + 1.5*IQR)))
df6 
```

```python
df6 = df6[df6['outlier']==True]
df6
```

```python
sns.boxplot(x=df6['Quantity'])
```

```python
# Chuẩn hóa dữ liệu
scaler  = MinMaxScaler()
df_s = scaler.fit_transform(df6[['Quantity','UnitPrice']])
pd.DataFrame(df_s).describe()
```

```python
# Vẽ lại biểu đồ
sns.boxplot(x=df_s)
```

```python
sns.kdeplot(df6['UnitPrice'])
```

```python
sns.kdeplot(data=df_s)
```
