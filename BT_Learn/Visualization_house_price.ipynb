```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("house_price.csv", encoding = "ISO-8859-1")
df.info()
```

```python
df.head()
```

```python
df = df.dropna(subset=['area','price'])
df    
```

```python
df = df[~((df['price']==0) | (df['area']== 0))]

df = df[~(df['area']== 0)]

df
```

+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

1. Biểu đồ mối liên hệ giữa diện tích với giá nhà

```python
plt.scatter(df['price'],df['area'],marker='*')
plt.show()
```

```python
plt.scatter(df['price'],df['bedroom'],marker='*')
plt.show()
```

```python
plt.scatter(df['price'],df['toilet'],marker='*')
plt.show()
```

```python
df['Unit_price']=df['price']/df['area']
df
```

```python
df1= df.groupby(['property_type']).mean().reset_index()
df1
```

```python
plt.bar(df1['property_type'],df1['Unit_price'])

plt.title('Gia trung bình', fontsize = 16)

plt.xlabel('Hình thức nhà', fontsize = 14)
plt.ylabel('Giá nhà', fontsize = 14)
plt.show()
```

```python
df2 = df.dropna(subset='bedroom')
df2
```
