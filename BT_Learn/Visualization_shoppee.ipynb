```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
df = pd.read_csv("shopee.csv", encoding = "ISO-8859-1")
df.info()
```

```python
df.head()
```

```python
df.shape
```

```python
df1 = df.drop_duplicates(subset='shopid')
df1.shape
```

```python
df1
```

```python
df2 = df1.groupby(['join_year']).count().reset_index()
df2
```

```python
plt.bar(df2['join_year'],df2['name'])

plt.title('Số lượng shop gia nhập theo các năm', fontsize = 16)

plt.xlabel('Year', fontsize = 14)

plt.ylabel('Count', fontsize = 14)

plt.show()
```

```python
df1['rating_good'] = df1['rating_bad']/(df1['rating_good']+df1['rating_normal'])
df1
```

```python

plt.scatter(df1['rating_good'],df1['response_rate']/100)

plt.title('mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt', fontsize = 16)

plt.xlabel('response_rate', fontsize = 14)
plt.ylabel('rating_good', fontsize = 14)
plt.show()
```

```python
plt.scatter(df1['response_time'],df1['rating_bad'])

plt.title('mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt', fontsize = 16)

plt.xlabel('response_time', fontsize = 14)

plt.ylabel('rating_bad', fontsize = 14)

plt.show()
```

```python
plt.plot(df2['join_year'],df2['name'])

plt.title('Xu hướng gia nhập theo thời gian', fontsize = 16)

plt.xlabel('Year', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.show()
```
