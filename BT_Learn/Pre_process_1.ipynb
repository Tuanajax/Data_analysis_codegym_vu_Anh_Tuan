```python
import pandas as pd
df = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv',encoding='ISO-8859-1')
# data.info()
df.info()
```

```python
df = df[['join_month', 'join_day','join_year','shop_location','rating_bad','rating_good','rating_normal']]
df
```

```python
# theo công thức sau: rating = rating_good *2 + rating_normal - rating_bad*3
df['rating']=df['rating_good']*2+df['rating_normal']-df['rating_bad']*3
df
```

```python
df['date'] = df['join_month'] + " " + df['join_day'].astype(str) + "," + df['join_year'].astype(str)
df
```

```python
df['new'] = df['join_year'] == 2021 
df
```

```python
import numpy as np
df['rate'] = np.where(df['rating_good'] >= 50000,'good','bad' )
df
```

```python
conditions = [(df['rating_good'] >= 30000) & (df['rating_bad'] <= 100),
              (df['rating_good'] >= 10000) & (df['rating_good'] < 30000) & (df['rating_bad'] <= 1000) & (df['rating_bad'] > 100),
              (df['rating_good'] < 10000)]
choices = ['blue', 'yellow','red']
df['flag'] = np.select(conditions, choices, default='black')
df 
```

```python
import pandas as pd
data = pd.read_csv('OnlineRetail.csv',encoding = 'unicode_escape')
data.info()
```

```python
data.head()
```

```python
# s.str.split(r"and|plus", expand=True)
mo_=data['InvoiceDate'].str.split(r"/|/").str[1]
_mo=[]
ls = list(mo_)
for x in ls:
    if int(x) < 4: _mo.append('1')
    elif int(x) < 7: _mo.append('2')
    elif int(x) < 10: _mo.append('3')
    else: _mo.append('4')
data['Quarter']=_mo
data.tail()
```

```python
data['Amount']= data['Quantity']*data['UnitPrice']
data
```

```python
conditions = [(data['Country'] == 'United Kingdom') & (data['Quarter'] == '4'),(data['Country'] == 'France')]
choices = [0.1,0.045]
data['Discount'] = np.select(conditions, choices, default=0)
data['Total'] = (1-data['Discount'].astype(float))*data['Amount']
data
```
