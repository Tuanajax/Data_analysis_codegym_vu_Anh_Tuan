```python
df_process = df.dropna(subset=['NumberOfDependents','MonthlyIncome'])
df_process.head()
```

Phát biểu:
+ H0: Thu nhập trung bình theo tháng của người phụ thuộc nhỏ hơn khách hành có phụ thuộc;
+ H1: Thu nhập trung bình theo tháng của người không phụ thuộc cao hơn có người phụ thuộc


```python
inpedent = df_process[df_process['NumberOfDependents'] ==0]['MonthlyIncome']

Depedent = df_process[df_process['NumberOfDependents'] > 0]['MonthlyIncome']

print(stats.ttest_ind(inpedent,Depedent,equal_var=False,nan_policy='omit'))
```
